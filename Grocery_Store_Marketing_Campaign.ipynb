```python
df.drop(columns='Year_Birth', axis=1, inplace=True)
```

```python
# get the total amount spent for each group
df["Total_Spent"] = df["MntWines"]+ df["MntFruits"]+ df["MntMeatProducts"]+ df["MntFishProducts"]+ df["MntSweetProducts"]+ df["MntGoldProds"]
```

```python
df["Total_Spent"]
```

```python
df["Children"]=df["Kidhome"]+df["Teenhome"]
df["Children"].head()
```

```python
df['Marital_Status'].unique()
```

```python
df['Adults'] = df['Marital_Status'].replace({'Married':2, 'Together':2, 'Single':1, 'Divorced':1, 'Widow':1, 'Alone':1, 'Absurd':1, 'YOLO':1})
df['Adults']
```

```python
df['Family_Total'] = df['Adults'] + df['Children']
```

```python
df = df.rename(columns={"MntWines": "Wines","MntFruits":"Fruits","MntMeatProducts":"Meat","MntFishProducts":"Fish","MntSweetProducts":"Sweets","MntGoldProds":"Gold"})
```

```python
df.drop(columns='Kidhome', axis=1, inplace=True)
```

```python
df.drop(columns='Teenhome', axis=1, inplace=True)
```

```python
# get the top groups of spenders 
# create a pie chart to see how the education levels 
# compare education levels to income levels 
# compare martial status to what they boought and see how much they spent 
# compare those with children to no children and what they spent most on 
```

```python
df.drop(columns=['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases',
       'NumStorePurchases', 'NumWebVisitsMonth', 'AcceptedCmp3',
       'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1', 'AcceptedCmp2',
       'Complain', 'Z_CostContact', 'Z_Revenue', 'Response'], axis=1, inplace=True)
```

```python
df.describe()
```

```python
education = df['Education'].value_counts()
```

```python
education.head()
```

```python
e = [0.0,0,0,0,0]
c = ["#DFFF00","#FFBF00","#FF7F50","#DE3163","#9FE2BF"]

plt.pie(education,
        labels=education.index,
        colors=c,
        explode=e,
        shadow=True,
        autopct="%.2f%%")
plt.title("Customer Education Levels");
# # plt.savefig("Education Levels.png")
```

```python
earning = df.groupby('Education')[['Income']].mean()
avg_income2 = earning.sort_values(by=['Income'], ascending = False)
```

```python
round(avg_income2,2)
```

```python
# plot average earned by each education level & average spent 

plt.barh(y=avg_income2.index, 
         width=avg_income2["Income"], 
         color=c)
plt.xlabel('Income');
plt.ylabel('Education Level');
plt.title("Average Income by Education");
# plt.savefig("Avg_Income_by_Edu.png")
plt.show()
```

```python
edu_spent = df.groupby('Education')[['Total_Spent']].mean()
edu_spent2 = edu_spent.sort_values(by=['Total_Spent'], ascending = False)
edu_spent2
```

```python
plt.barh(y=edu_spent2.index, 
         width=edu_spent2["Total_Spent"], 
         color=c)
plt.xlabel('Average Spent');
plt.ylabel('Education Level');
plt.title("Average Spent by Education Level");
# plt.savefig("Avg_Income_by_Edu.png")
plt.show()
```
